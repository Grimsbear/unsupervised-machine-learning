```python
#Imports
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#Read-In
file_path = "Resources/myopia.csv"
starter_df = pd.read_csv(file_path)
starter_df.head(5)
```

```python
#Clean-Up
myopia_df = starter_df.drop("MYOPIC",axis = 1)
myopia_df
```

```python
#Scaling
myopia_scaled = StandardScaler().fit_transform(myopia_df)
myopia_scaled
```

```python
#PCA Dimension Reduction
pca = PCA(n_components=0.90)

myopia_pca = pca.fit_transform(myopia_scaled)
myopia_pca
```

```python
#t-SNE Dimension Reduction
tsne = TSNE(learning_rate="auto")

myopia_tsne = tsne.fit_transform(myopia_pca)
```

```python
#Checking Shape
myopia_tsne.shape
```

```python
#Plotting Data
myopia_df["X"] = myopia_tsne[:,0]
myopia_df["y"] = myopia_tsne[:,1]

plt.scatter(myopia_df['X'], myopia_df['y'])
plt.show()
```

```python
#Testing for K-Means
inertia = []

k = list(range(1, 11))

for i in k:
    km = KMeans(n_clusters=i, random_state=18)
    km.fit(myopia_df)
    inertia.append(km.inertia_)

elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

# Final Conclusion

### I believe that the K-Means Elbow for Inertia is at 3, Inertia won't drop another 100,000 from 3 til somewhere around 8. Based on the data above I would say that we might find 3 groups.
